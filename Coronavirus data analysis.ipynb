```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# GET THE DATA FROM THE URL GIVEN IN THE Johns Hopkins University REPOSITORY: https://github.com/CSSEGISandData

CONFIRMED_CASES_TIME_SERIES_CSV = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
DEATH_CASES_TIME_SERIES_CSV = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
RECOVERED_CASES_TIME_SERIES_CSV = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'




```

# Read the data in Pandas and save it into a variable




```python

confirmed_cases_timeseries = pd.read_csv(CONFIRMED_CASES_TIME_SERIES_CSV)
death_cases_timeseries = pd.read_csv(DEATH_CASES_TIME_SERIES_CSV)
recovered_cases_timeseries = pd.read_csv(RECOVERED_CASES_TIME_SERIES_CSV)
```

# Export the CSV files

```python
confirmed_cases_timeseries.to_csv("confirmed_cases_timeseries"+'.csv')
confirmed_cases_timeseries
```

# Export the CSV files

```python
death_cases_timeseries.to_csv("death_cases_timeseries"+'.csv')
death_cases_timeseries
```

# Export the CSV files

```python
recovered_cases_timeseries.to_csv("recovered_cases_timeseries"+'.csv')

recovered_cases_timeseries
```

# Sum of the number of cases per day in total:

```python

all_countries_confirmed_cases_timeseries = confirmed_cases_timeseries.sum()
all_countries_death_cases_timeseries = death_cases_timeseries.sum()
all_countries_recovered_cases_timeseries = recovered_cases_timeseries.sum()

all_countries_confirmed_cases_timeseries
```

# Export to a CSV file

```python
all_countries_confirmed_cases_timeseries.to_csv("all_countries_confirmed_cases_timeseries"+'.csv')
all_countries_death_cases_timeseries.to_csv("all_countries_death_cases_timeseries"+'.csv')
all_countries_recovered_cases_timeseries.to_csv("all_countries_recovered_cases_timeseries"+'.csv')

```

# Sum of the cases equals to "Germany" in the column "country/region", and then export to a CSV file:

```python
germany_confirmed_cases_timeseries = confirmed_cases_timeseries[confirmed_cases_timeseries['Country/Region']=='Germany'].sum()
germany_death_cases_timeseries = death_cases_timeseries[death_cases_timeseries['Country/Region']=='Germany'].sum()
germany_recovered_cases_timeseries = recovered_cases_timeseries[recovered_cases_timeseries['Country/Region']=='Germany'].sum()
```
